# Приручник за Џупајтер (_Jupyter_) 

Пред вама је Приручник за Џупајтер (_Jupyter_), интерактивно радно окружење за извршавање Пајтон програма. У њему је обрађен део курса информатике и рачунарства за 8. разред основне школе у Србији који се односи на визуелизацију и обраду података.

Сваки час је обрађен у једној Џупајтер радној свесци. Први час представља увод у Џупајтер и мало подсећање на основе програмског језика Пајтон, док је осталих петнаест часова подељено у групе од по три часа (два часа обраде градива и један час утврђивања градива).

Ако желите да баците поглед како свеске изгледају унутар Џупајтера, можете отићи на  
https://mybinder.org/v2/gh/Petlja/JupyterPrirucnik/master?filepath=Sadrzaj.ipynb. 

Свеске прилагођене интерактивном раду можете преузети са _GitHub_ репотиторијума https://github.com/Petlja/JupyterPrirucnik и отворити их у локалној инсталацији Џупајтера или коришћењем сервиса у облаку као што је [Azure Notebooks](https://notebooks.azure.com/). _Azure Notebooks_ има опцију да креирате нову библиотеку (_Library_) на основу _GitHub_ репозиторијума (користећи горе наведенy адресу репозиторијума).
